````python
# Exercise 2) Load all the important library
import pandas
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.neighbors import KNeighborsClassifier
````

````python
# Exercise 3) Load the iris dataset from file (iris.data)
names = ['sepal - length', 'sepal - width', 'petal - length', 'petal - width', 'class'] # set the name of each column
ds = pandas.read_csv('iris.data', names = names) # read the dataset, and put name into each column
# display the dataset
ds
````

````python
# Exercise 4) Display basic information of this dataset
print(ds.shape)     # display rows x columns of the dataset
print(ds.head(20))  # print first 20 rows of the dataset
print(ds.describe())  # show number based on the statistic of the dataset (Ex: min, max)
print(ds.groupby('class').size()) # group all of the items on the dataset by using 'class' to seperate
````

````python
# Exercise 5) Create a box plots on iris dataset (Lesson from week 7)
ds.plot(kind = 'box', subplots = True, layout = (2,2), # layout = (width,height)
        sharex = False, sharey = False) # Controls sharing of properties among x (sharex) or y (sharey) axes:
plt.show()
````

--- Milestone 1 Completed ---

````python
# Exercise 6) Seperate features of iris flower from the target
array = ds.values   # returns a list of all the values in the dataset
X = array[:,0:4]    # get all items (:), from columns 0-3 (feature)
Y = array[:,4]      # get all items, from column 4 (target)
print("X size:", X.size)
print("Y size:", Y.size)

print(X)
print(Y)
````

````python
# Exercise 7) Split iris.data into two groups (For training and testing)
test_size = 0.20
seed = 7
X_train , X_test , Y_train , Y_test = model_selection.train_test_split(X, Y, test_size = test_size, random_state = seed)
# used to split the dataset,
# based on test_size(size of the testing dataset) and
# random_state(Split by using seed's size, if not it's going to random split by using np.random)
print("X_train_size:", X_train.size)
print("X_test_size:", X_test.size)
print("Y_train_size:", Y_train.size)
print("Y_test_size:", Y_test.size)

print("X_train:\n", X_train)
print("X_test:\n", X_test)
print("Y_train:\n", Y_train)
print("Y_test:\n", Y_test)
````

````python
# Exercise 8) Change num 7
test_size = 0.5
seed = 10
X_train , X_test , Y_train , Y_test = model_selection.train_test_split(X, Y, test_size = test_size, random_state = seed)
# used to split the dataset, based on
# test_size(size of the testing dataset)
# and random_state(Split by using seed's size(seed of data), if not it's going to random split by using np.random)
print("X_train_size:", X_train.size)
print("X_test_size:", X_test.size)
print("Y_train_size:", Y_train.size)
print("Y_test_size:", Y_test.size)

print("X_train:\n", X_train)
print("X_test:\n", X_test)
print("Y_train:\n", Y_train)
print("Y_test:\n", Y_test)
````

### __K-nearest neighbors(KNN)__ is a type of supervised learning algorithm used for both regression and classification. KNN tries to predict the correct class for the test data by calculating the distance between the test data and all the training points.

Exercise 9) The K-nearest Neighbors state that the new object will be classified to the majority vote of it k-nearest neighbors. Hence, if we considered

__ANS:__ I think ```k=3``` is more reliable, since the finding range is bigger and as of the result it makes the test data matches with the catagories that it should be/looks like the most.

````python
# Exercise 10) Using K-nearest neighbors technique, now use 7 neighbors
irisML = KNeighborsClassifier(n_neighbors = 7)
irisML.fit(X_train, Y_train)
````

--- Milestone 2 Completed ---

````python
# Exercise 11) Pass the test dataset to the model for the prediction
prediction = irisML.predict(X_test)
print(prediction) #.predict(): predict the labels of the data values on the basis of the trained model
print(prediction.size)
````

````python
# Exercise 12) To see the score of prediction model
prediction = irisML.score(X_test, Y_test)
print(prediction) # .score(): the method computes the accuracy score by default of the model
````

````python
# Exercise 13) Input new iris object to the model to classify its specie
new_obj1 = [[5.2 ,4.3 ,5.6 ,1.6]]
output1 = irisML.predict(new_obj1)
print(output1)

new_obj2 = [[3.2, 4.1, 2.1, 1.0]]
output2 = irisML.predict(new_obj2)
print(output2)
````
