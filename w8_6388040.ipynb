```python
# Exercise 1
import matplotlib.pyplot as plt
# dot plot
plt.plot(1, 3, 'bo')
plt.plot(2, 2, 'ro')
plt.plot(1.4, 2.4, 'go')

# line plot
plt.plot([1, 3, 12, 30]) # blue : this is value in y-axis within each second (0, 1, 2, and so on)
plt.plot([1, 4, 3, 12])  # orange

plt.plot([0, 3, 5, 6], [0, 5, 15, 10]) # green : plot x and y like (0, 0), (3, 5), ...
plt.plot([0, 1, 2, 6], [10, 8, 1, 3]) # red
```

```python
# Exercise 2 PLot Decoration
# - normal line, -- dashed line, : dotted line (like sqaure)
# -. dashed line with dot
plt.plot ([0, 3, 5, 6] , [0, 5, 15, 10], '--',label ='A',
           color = "pink", linewidth = 5)
plt.plot ([0 , 1 , 2 , 6] , [10 , 8 , 1 , 3], ':',label ='B',
           color = "lightblue", linewidth = 4)

plt.legend () # Show little A and B border

fig = plt.figure(figsize=(6,2)) # figsize : adjust your graph (w * h)
plt.plot([0, 3, 5, 6] , [0, 5, 15, 10], '--',label ='A',
           color = "pink", linewidth = 5)
plt.plot([0 , 1 , 2 , 6] , [10 , 8 , 1 , 3], ':',label ='B',
           color = "lightblue", linewidth = 2)
```

```python
# Exercise 3 Bar Chart
import matplotlib.pyplot as plt
plt.barh([1, 2, 3] , [0, 1, 5], color = 'orange')
# x and y along the graph (x, y) => (1, 0), (2, 1), (3, 5)
```

```python
import numpy as np
n = 6
x = np.arange(n)
y1 = (1 - x / float(n))
y2 = (1 - x / float(n))
plt.bar(x, +y1 , facecolor ='blue', edgecolor ='yellow')
plt.bar(x, -y2 , facecolor ='red', edgecolor ='green')
```

```python
plt.barh([1, 2, 3] , [0, 1, 5]) # blue (2, 1) overlap under (2, 1) orange
plt.barh([2, 2, 5] , [1, 1, 1]) # orange
plt.bar([1, 2, 3] , [3, 4, 5])  # green
```

```python
# Exercise 4 subplot
fig = plt.figure()
ax = fig.add_subplot(111)  # 111 : refers to the number of rows, the number of columns, and the plot number.
ax.barh([1, 2, 3] , [0, 1, 5]) # blue
fig = plt.figure(figsize = (20, 10)) # Set size of graph (w * h)
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)
ax1.barh([1, 2, 3] , [0, 1, 5], color = 'red')
ax2.barh([2, 2, 5] , [1, 1, 1], color = 'green')
ax3.bar([1, 2, 3] , [3, 4, 5], color = 'blue')
```

```python
# Exercise 5 Stack plot
rng = np.arange(50)
rnd = np.random.randint(0, 10, size = (3, rng.size)) # random number between 0 to 10 # make it 3 list, and accroding to rng size
yrs = 1950 + rng
fig, ax = plt.subplots(figsize = (5, 3))
ax.stackplot(yrs, rng + rnd,
              labels =['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time') # Set title
ax.legend(loc = 'upper left') # Set the white border (Description)
ax.set_ylabel('Total debt') # Set y label text
ax.set_xlim(xmin = yrs[0], xmax = yrs [-1]) # Set the x-axis view limits
fig.tight_layout() #  automatically adjust subplot parameters to give specified padding.
```

```python
import seaborn as sns
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x = "total_bill", y = "tip", hue = "smoker",     # hue : Grouping variable that will produce elements with different colors. 
            style = "smoker", size = "size", data = tips)    # style : Grouping variable that will produce elements with different styles. 
                                                             # size : Grouping variable that will produce elements with different sizes.  
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x = "total_bill", y = "tip", col = "time", hue ="smoker", # col : Variables that define subsets to plot on different facets. (time / dinner)
              style = "smoker", size = "size", data = tips)
```

```python
sns.relplot(x = "total_bill", y = "tip", col ="time", hue = "size",
            style = "smoker", size = "size", palette = "YlGnBu", # palette : Method for choosing the colors to use when mapping the hue semantic.
            markers = ["D", "o"], sizes = (10, 125), edgecolor =".2", # markers : Object determining how to draw the markers for different levels of the style variable.
            linewidth = .5, alpha = .75, data = tips)           # alpha : To control the transparency of plots

sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x = "total_bill", y = "tip", col = "smoker", hue = "sex",
            kind = "line", data = tips)
tips.head() # head : Return the first n rows. (pandas library)
```

```python
# Exercise 10 : Categorical plots (catplot = simplifies plotting that involves categorical variables.)
import seaborn as sns
sns.set()
tips = sns.load_dataset("tips")

# sns.catplot(x = "day", y = "total_bill", hue = "smoker",
#            data = tips)

sns.catplot(x = "day", y = "total_bill", hue = "smoker",
            kind = "swarm", data = tips)

sns.catplot(x = "day", y = "total_bill", hue = "smoker",
            kind = "violin", split = True , data = tips)

sns.catplot(x ="day", y = "total_bill", hue = "smoker",
            kind = "bar", data = tips)      # little black line above bar -> refers to max and min value
```
